```python
import pandas as pd
file=pd.read_csv("/content/IRIS.csv")
x=file[["sepal_length","sepal_width","petal_length","petal_width",]][0:50]
from sklearn.decomposition import PCA
pcamodel=PCA(n_components=2)
principalcomponents=pcamodel.fit_transform(x)
newx=pd.DataFrame(data=principalcomponents,columns=["PC1","PC2"])
xaxis=newx["PC1"]
yaxis=newx["PC2"]
newx
```

```python
from scipy.cluster import hierarchy
```

```python
import matplotlib.pyplot as plt
```

```python
Z = hierarchy.linkage(x, 'single')
```

```python
plt.figure()
```

```python
hierarchy.dendrogram(Z, above_threshold_color='green',
                           orientation='top')
```
