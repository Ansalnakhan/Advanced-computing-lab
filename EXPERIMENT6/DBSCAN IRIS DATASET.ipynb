```python
import pandas as pd
file=pd.read_csv("/content/IRIS.csv")
x=file[["sepal_length",	"sepal_width", "petal_length",	"petal_width"]][0:50]
from sklearn.decomposition import PCA
pcamodel=PCA(n_components=2)
principalcomponents=pcamodel.fit_transform(x)
newx=pd.DataFrame(data=principalcomponents,columns=["PC1","PC2"])
xaxis=newx["PC1"]
yaxis=newx["PC2"]
newx
```

```python
from sklearn.cluster import DBSCAN
```

```python
ML=DBSCAN()
```

```python
ML.fit(newx[:])
```

```python
labels=ML.labels_
```

```python
import matplotlib.pyplot as plt
```

```python
xaxis=newx["PC1"]
```

```python
yaxis=newx["PC2"]
```

```python
plt.scatter(xaxis,yaxis,c=labels,cmap="rainbow")
```
