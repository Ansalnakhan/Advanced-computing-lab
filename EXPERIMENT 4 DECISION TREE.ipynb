```python
import pandas as pd
file=pd.read_csv("/content/bank.csv")
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
z=["job", "marital", "education", "default", "housing",	"loan",	"contact", "month"]
```

```python
file[z]=file[z].apply(le.fit_transform)
```

```python
file
```

```python
X=file[["job", "marital", "education", "default", "housing",	"loan",	"contact", "month"]]
Y=file["deposit"]
X
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(random_state=0)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,train_size=0.7)
```

```python
clf=clf.fit(xtrain,ytrain)
```

```python
res=clf.predict(xtrain)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytrain, res)
```

```python
from sklearn import tree
```

```python
import graphviz
```

```python
dot_data = tree.export_graphviz(clf)
```
